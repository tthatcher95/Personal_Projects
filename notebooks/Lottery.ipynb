```python
import urllib2
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import *
```

```python
def read_data():
    x = ["Draw #", 'Draw Date1', 'Draw Date2', 'Draw Date3','Draw Date4', 'Number_1', 'Number_2', 'Number_3', 'Number_4', 'Number_5', 'Powerball']
    df = pd.read_csv("DownloadAllNumbers.txt", delim_whitespace=True, skiprows=1, names=x)
    df['Date'] = df['Draw Date1'] + ' ' + df['Draw Date2'] + ' ' + df['Draw Date3'] + ' '+ df['Draw Date4']
    df['Year'] = df['Draw Date4']
    df['Month'] = df['Draw Date2']
    df = df[1:]
    df
    df = df.drop(columns=['Draw Date1', 'Draw Date2', 'Draw Date3', 'Draw Date4'])
    df['Number_3'] = pd.to_numeric(df['Number_3'])
    df['Number_2'] = pd.to_numeric(df['Number_2'])
    df['Number_1'] = pd.to_numeric(df['Number_1'])
    return df

data_df = read_data()
data_df

```

```python
# data = loadtxt('data.txt',dtype=int,usecols=(4,)) #loading 5th column of csv file into array named data. 
# data = df['Powerball']
# plt.hist(data, bins=np.arange(data.min(), data.max()+1))
# df_top_freq = df.groupby(['Date', 'Draw #', 'Powerball'])['Powerball'].agg(
#     {"code_count": len}).sort_values(
#     "code_count", ascending=True).head(10).reset_index()

def best_guess_df(df, n=10, year=None, month=None):
    
    def filter_df(df, year=None, month=None):
#         if(year and month):
#             return df.loc[df['Year'] == str(year) and df['Month'] == str(month)]
        
        if(year):
            return df.loc[df['Year'] == str(year)]

        if(month):
            return df.loc[df['Month'] == str(month)]

        else:
            return df

    new_df = filter_df(df=df, year=year, month=month)

    def get_top(df, colname="Number_1", n=n):
        numdf = pd.DataFrame(index=range(0, n))
        countdf = df[colname].value_counts().nlargest(n)
        numdf[colname] = zip(countdf.index.values, countdf)
        return numdf, len(df[colname])

    num1, n1 = get_top(new_df, "Number_1")
    num2, n2 = get_top(new_df, "Number_2")
    num3, n3 = get_top(new_df, "Number_3")
    num4, n4 = get_top(new_df, "Number_4")
    num5, n5 = get_top(new_df, "Number_5")
    powerball, p = get_top(new_df, "Powerball")

    num_df = pd.DataFrame(columns=["Number_1", "Number_2", "Number_3", "Number_4", "Number_5", "Powerball"])

    num_df["Number_1"] = num1['Number_1']
    num_df["Number_2"] = num2['Number_2']
    num_df["Number_3"] = num3['Number_3']
    num_df["Number_4"] = num4['Number_4']
    num_df["Number_5"] = num5['Number_5']
    num_df["Powerball"] = powerball['Powerball']

    def best_guess(df):
        return str(df['Number_1'][0]) + ' ' + str(df['Number_2'][0]) + ' ' + str(df['Number_3'][0]) + ' ' + str(df['Number_4'][0]) + ' ' + str(df['Number_5'][0]) + ' ' + str(df['Powerball'][0])

    num_df['Best_Guess'] = num_df.apply(best_guess, axis=1)
    
    return num_df

best_df = best_guess_df(data_df[1:], n=20, year=2019, month='May')
best_df
```
